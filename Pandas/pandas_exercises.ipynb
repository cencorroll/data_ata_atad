```python
# Question 1
import pandas as pd
data = {
    'Name': ['John', 'Mary', 'Alex', 'Sarah', 'Jack'],
    'Math_Score': [85, 78, 92, 95, 80],
    'Science_Score': [90, 92, 88, 89, 85],
    'History_Score': [88, 85, 91, 92, 86]
}
df = pd.DataFrame(data)
df
```

```python
# Task A
df['Name'].count()
```

```python
# Task B
amount = df['Name'].count()
avg_math_score = sum(df['Math_Score']) / amount
avg_science_score = sum(df['Science_Score']) / amount
avg_history_score = sum(df['History_Score']) / amount

avg_scores_df = pd.DataFrame({
    'Subject': ['Maths', 'Science', 'History'],
    'Average Score': [avg_math_score, avg_science_score, avg_history_score]
})

avg_scores_df
```

```python
# Task C
total_scores = df['Math_Score'] + df['Science_Score'] + df['History_Score']

total_scores_df = pd.DataFrame({
    'Name': df['Name'],
    'Total Scores': total_scores
})

highest_score = total_scores_df['Total Scores'].max()
student_with_highest_score = total_scores_df[total_scores_df['Total Scores'] == highest_score]
student_with_highest_score
```

```python
# Question 2
data = {
    'Date': ['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04', '2023-07-05'],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone'],
    'Sales': [1000, 500, 800, 1200, 600],
    'Profit': [200, 100, 150, 250, 120]
}

df = pd.DataFrame(data)
df
```

```python
# Task A
total_sales = df.groupby(['Product'])['Sales'].sum()
total_profit = df.groupby(['Product'])['Profit'].sum()
product_totals = pd.DataFrame({
    'Product': total_sales.index,
    'Total Sales': total_sales.values,
    'Total Profit': total_profit.values
})
product_totals
```

```python
highest_total_sales = product_totals['Total Sales'].max()
highest_total_profit = product_totals['Total Profit'].max()

product_with_highest_sales = product_totals[product_totals['Total Sales'] == highest_total_sales]
product_with_highest_sales

product_with_highest_profit = product_totals[product_totals['Total Profit'] == highest_total_profit]
product_with_highest_profit
```

```python
profit_percentage = (df['Profit'] / df['Sales']) * 100
profit_percentage = round(profit_percentage, 2)
df['Profit Percentage'] = profit_percentage
df
```

```python
# Question 3
data = {
    'Employee_ID': [1001, 1002, 1003, 1004, 1005],
    'Name': ['John', 'Sarah', 'Alex', 'Emily', 'Jack'],
    'Department': ['Marketing', 'Accounting', 'IT', 'HR', 'Marketing'],
    'Salary': [50000, 60000, 55000, 58000, 52000]
}

df = pd.DataFrame(data)
df
```

```python
# Task A
avg_salary_per_department = df.groupby('Department')['Salary'].mean()

avg_salary_per_department_df = pd.DataFrame(avg_salary_per_department).reset_index()
avg_salary_per_department_df
```

```python
# Task B
highest_salary = avg_salary_per_department_df['Salary'].max()
department_with_highest_salary = avg_salary_per_department_df[avg_salary_per_department_df['Salary'] == highest_salary]
department_with_highest_salary
```

```python
# Task C

#df['Salary Category'] = df['Salary'].apply(lambda x: 'High' if x > avg_salary_per_department else 'Low')
df['Salary Category'] = df.apply(lambda row: 'High' if row['Salary'] > avg_salary_per_department[row['Department']] else 'Low', axis=1)

df
```

```python
# Question 4
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'West'],
    'Sales': [1000, 800, 1200, 900, 1100, 750],
    'Profit': [200, 150, 250, 180, 220, 130]
}

df = pd.DataFrame(data)
df
```

```python
# Task A
total_sales = df.groupby(['Region'])['Sales'].sum()
total_profit = df.groupby(['Region'])['Profit'].sum()
region_totals = pd.DataFrame({
    'Region': total_sales.index,
    'Total Sales': total_sales.values,
    'Total Profit': total_profit.values
})
region_totals
```

```python
# Task B
highest_total_sales = region_totals['Total Sales'].max()
highest_total_profit = region_totals['Total Profit'].max()

region_with_highest_sales = region_totals[region_totals['Total Sales'] == highest_total_sales]
region_with_highest_sales

region_with_highest_profit = region_totals[region_totals['Total Profit'] == highest_total_profit]
region_with_highest_profit
```
